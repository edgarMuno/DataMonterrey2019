```python
import pandas as pd
import os

# Lista con los nombres de los archivos de Excel de los 12 meses (ajusta los nombres segÃºn sea necesario)
archivos_excel = [
    'ENERO 2019.xlsx', 'FEBRERO 2019.xlsx', 'MARZO 2019.xlsx',
    'ABRIL 2019.xlsx', 'MAYO 2019.xlsx', 'JUNIO 2019.xlsx',
    'JULIO 2019.xlsx', 'AGOSTO 2019.xlsx', 'SEPTIEMBRE 2019.xlsx',
    'OCTUBRE 2019.xlsx', 'NOVIEMBRE 2019.xlsx', 'DICIEMBRE 2019.xlsx'
]

# Crear una lista para almacenar los DataFrames de todos los meses
lista_dataframes = []

# Iterar sobre cada archivo de Excel en la lista
for archivo_excel in archivos_excel:
    # Leer todas las hojas del archivo Excel
    hojas = pd.read_excel(archivo_excel, sheet_name=None, skiprows=5)
    
    # Iterar sobre las hojas de cada archivo
    for nombre_hoja, df in hojas.items():
        if nombre_hoja != 'Inicio':  # Excluir la hoja "Inicio" si no es necesaria
            df = df.rename(columns={'FACT': 'FACT-NUE-FO'})
            df = df.rename(columns={'FECHA\nPAGO': 'FECHA'})
      
            #df = df.rename(columns={'FDO': 'FDO - #OP'})
            # AÃ±adir el nombre de la hoja como una nueva columna
            df['Hoja'] = nombre_hoja
            # AÃ±adir el nombre del archivo o mes como una nueva columna
            df['Mes'] = os.path.basename(archivo_excel).split(' ')[0]  # Extraer el mes del nombre del archivo
            # Agregar el DataFrame a la lista
            lista_dataframes.append(df)

# Concatenar todos los DataFrames en uno solo
df_final = pd.concat(lista_dataframes, ignore_index=True)

# Mostrar el DataFrame final
df_final
```
