```python
import pandas as pd
import matplotlib.pyplot as plt

import matplotlib.ticker as ticker
```

```python
df= pd.read_csv('compras_2019.csv')
df
```

```python
df.dtypes
```

```python
df['FECHA']=pd.to_datetime(df['FECHA'])
```

```python
df['DIA']=df['FECHA'].apply(lambda x:x.day)
```

```python
df['MES_NUM']=df['FECHA'].apply(lambda x:x.month)
```

```python
df['ANIO']=df['FECHA'].apply(lambda x:x.year)
```

## Proyectar una variable con otra

### Numero de trasacciones por mes

```python

Mes_trans=df['Mes'].value_counts().to_dict()
```

```python
list(Mes_trans.keys())[0:5]
```

```python
df['Num_trans']=df['Mes'].apply(lambda x: Mes_trans[x])
```

```python
df.head(3)
```

### Promedio de gastos por mes

```python
promedio_mes=df.groupby('Mes')['TOTAL_EGRESO'].mean()
```

```python
list(promedio_mes.keys())[0:5]
```

```python
df['GASTO_PROMEDIO_MES']=df['Mes'].apply(lambda x:promedio_mes[x])
```

### Numero de transacciones por Beneficiario

```python
beneficiario_sales=df['BENEFICIARIO'].value_counts().to_dict()
```

```python
list(beneficiario_sales.keys())[0:5]
```

```python
df['BENEFICIARIO_SALES']=df['BENEFICIARIO'].apply(lambda x: beneficiario_sales[x])
```

```python
df.head(3)
```

### Total de gastos por Hoja

```python
promedio_hoja=df.groupby('Hoja')['TOTAL_EGRESO'].mean()
```

```python
promedio_hoja.to_dict()
```

```python
df['GASTO_PROMEDIO_HOJA']=df['Hoja'].apply(lambda x:promedio_hoja[x])
```

```python
df.head(3)
```

### Exportar a CSV

```python
df.to_csv('compras_2019_enrequecida.csv',index=None)
```
